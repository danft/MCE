```python
from sympy import *

#from math import cos, sin, pi, sqrt, tan, fabs
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize

init_printing()

%matplotlib inline
from mpmath import *
```

```python
m=Symbol('m')
a=Symbol('a')
b=Symbol('b')
x1=Symbol('x1')
y1=Symbol('y1')
x2=Symbol('x2')
y2=Symbol('y2')
D=Symbol('D')
t=Symbol('t')

C = Symbol('C')
S = Symbol('S')

q = b/a
e = Symbol('e')

C = (e**(j * t) + e**(-j*t))/2
S = (e**(j * t) - e**(-j*t))/(2*j)

C = (e+ 1/e)/2
S = (e-1/e)/(2j)

def px(x, y):
    return q * (x * C + y * S)

def py(x, y):
    return -x * S + y * C

px1 = px(x1, y1)
px2 = px(x2, y2)
py1 = py(x1, y1)
py2 = py(x2, y2)

leu = px1**2 + py1**2
lev = px2**2 + py2**2
leuv = (px1-px2)**2 + (py1-py2)**2

A = (px1*py2 - py1*px2)**2/4

xi = 16 * b**2 * A - leu*lev*leuv
expand(xi)
```

```python
p = Poly(e ** 6 * xi, [e])
```

```python
dir(p)
```

```python
simplify(p.coeff_monomial(e**12))
```
