```python
from sympy import *

#from math import cos, sin, pi, sqrt, tan, fabs
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
from fractions import Fraction

#init_printing()

%matplotlib inline
from mpmath import *
```

```python
m=Symbol('m', real=True)
a=Symbol('a', real=True)
b=Symbol('b', real=True)
x1=Symbol('x1', real=True)
y1=Symbol('y1', real=True)
x2=Symbol('x2', real=True)
y2=Symbol('y2', real=True)
D=Symbol('D', real=True)
t=Symbol('t', real=True)

C = Symbol('C', real=True)
S = Symbol('S', real=True)

q = b/a
e = Symbol('e', real=True)

C = (e**(j * t) + e**(-j*t))/2
S = (e**(j * t) - e**(-j*t))/(2*j)

C = (e+ 1/e)/2
S = (e-1/e)/(2j)

def px(x, y):
    return q * (x * C + y * S)

def py(x, y):
    return -x * S + y * C

px1 = px(x1, y1)
px2 = px(x2, y2)
py1 = py(x1, y1)
py2 = py(x2, y2)

leu = px1**2 + py1**2
lev = px2**2 + py2**2
leuv = (px1-px2)**2 + (py1-py2)**2

A = (px1*py2 - py1*px2)**2/4

xi = 16 * b**2 * A - leu*lev*leuv
#expand(xi)
```

```python
p = Poly(e ** 6 * xi, [e])
```

```python
def plat(cdic):
    
    cnt = 0
    expr = 0
    
    for  key in cdic:
        cnt += 1
        nexpr = simplify(expr + key * cdic[key])
        
        nss = latex(nexpr)
        if len(nss) - nss.count(' ') - 4*nss.count('left') - 5*nss.count('right') - nss.count('{') - nss.count('}') - nss.count('\\') - 4*nss.count('frac') - nss.count('_') - nss.count('^') > 92:
            nss = latex(expr)
            if len(nss) > 0:
                #print("BLA")
                if nss[0]!='-':
                    print(f"$$+{nss}$$")
                else:
                    print(f"$${nss}$$")

            expr = simplify(key * cdic[key])
        else:
            expr = nexpr
        
        #expr += key * cdic[key]

        #if cnt % 5 == 0:
        #  print(f"{latex(expr)}\\\\")
        # expr = 0
            
    
    print(f"$${latex(expr)}$$")
    
    
#k12 = expand(simplify(p.coeff_monomial(e**0)))
k12 = p.coeff_monomial(e**12)

#dir(k12)
plat(k12.as_coefficients_dict())
```
